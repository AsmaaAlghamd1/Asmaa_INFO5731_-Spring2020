```python

MyFile = open('AdamsTanner.txt','r', encoding='utf8', errors = "ignore")
lines = MyFile.readlines()

MyFile2 = []
for line in lines:
  nlp = spacy.blank('en')
  nlp.add_pipe(PySBDFactory(nlp))
  doc = nlp(line)
  for item in list(doc.sents):
    MyFile2.append(item)

MyFile3 = []
for sent in MyFile2:
  z = lexnlp.extract.en.amounts.get_amounts(sent)
  MyFile3.append(z)

print(MyFile3)
```
