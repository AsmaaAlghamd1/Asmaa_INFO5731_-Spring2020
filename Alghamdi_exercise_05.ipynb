```python
def total(Pos_Tag):
  Adj = 0
  Adv = 0
  Nou=0
  Ver=0
  For_Word=0
  Nou_Plu=0
  Determ = 0
  Cord_Conj = 0
  Sub_Conj =0
  Int_jec = 0
  Pers_Pron = 0
  Predet = 0
 

  Adjective =[]
  Adverb = []
  Noun = []
  Verb = []
  Foreign_Word=[]
  Noun_Plural=[]
  Determiner = []
  Cordinating_Conjunction = []
  Subordinating_Conjuction=[]
  Interjection = []
  Personal_Pronoun = []
  Predeterminer = []
  
  
  
  for i in Pos_Tag:
    for j in i:
      if (j[1]=='JJ'):
        Adj+=1
      elif (j[1]=='RB'):
        Adv+=1
      elif (j[1]=='NN'):
        Nou+=1
      elif (j[1]=='VR'):
        Ver+=1
      elif (j[1]=='FW'):
        For_Word+=1
      elif (j[1]=='NNS'):
        Nou_Plu+=1
      elif (j[1]=='DT'):
        Determ +=1
      elif (j[1]=='CC'):
        Cord_Conj+=1
      elif (j[1]=='IN'):
        Sub_Conj+=1
      elif (j[1]=='UH'):
        Int_jec+=1
      elif (j[1]=='PRP'):
        Pers_Pron+=1
      elif (j[1]=='PDT'):
        Predet+=1
      else :0
  Adjective.append(Adj)
  Adverb.append(Adv)
  Noun.append(Nou)
  Verb.append(Ver)
  Foreign_Word.append(For_Word)
  Noun_Plural.append(Nou_Plu)
  Determiner.append(Determ)
  Cordinating_Conjunction.append(Cord_Conj)
  Subordinating_Conjuction.append(Sub_Conj)
  Interjection.append(Int_jec)
  Personal_Pronoun.append(Pers_Pron)
  Predeterminer.append(Predet)
  
  return Adjective,Adverb,Noun,Verb,Foreign_Word,Noun_Plural,Determiner,Cordinating_Conjunction,Subordinating_Conjuction,Interjection,Personal_Pronoun,Predeterminer
```
