# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(5) Stemming. 

(6) Lemmatization.

```python
d = pd.read_csv("List_tweets.csv",sep='\t',names=['tweets'])
import string

d['Remove Noise'] = d['tweets'].apply(lambda x: " ".join(x for x in x.split() if x.isalnum()))
d[['Remove Noise','tweets']].head(n=100)
```

```python



# Remove special characters
d['Special Character Remover'] = d['Remove Noise'].str.replace("[^a-zA-Z#]+", " ").str.strip()
d[['Special Character Remover','Remove Noise']].head(n=100)


```

```python
# Remove numbers
d["Remove numbers"] = d['Special Character Remover'].str.rstrip(string.digits)
d[["Remove numbers",'Special Character Remover']].head(n=100)

```

```python

#lower casing
d['lower casing']=d["Remove numbers"].apply(lambda x: " ".join(x.lower() for x in x.split()))
d[["Remove numbers",'lower casing']].head(n=100)


```

```python
#Remove stop words
from nltk.corpus import stopwords
nltk.download('stopwords')
stop=stopwords.words('english')

d['Stop words removal']=d['lower casing'].apply(lambda x:" ".join(x for x in x.split() if x not in stop))
d[['Stop words removal','lower casing']].head(n=100)


```

```python
#Remove punctuations

d['Punctuations removal']=d['Stop words removal'].str.replace('[^\w\s]','')
d[['Punctuations removal','Stop words removal']].head(n=100)

```

```python
#lemmatization

nltk.download('wordnet')
d['lemmatization']=d['Punctuations removal'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
d[['lemmatization','Punctuations removal']].head(n=100)
```

```python

#stemming

st=PorterStemmer()

d['Stemming']=d['lemmatization'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
d[['Stemming','lemmatization']].head(n=100)
```
